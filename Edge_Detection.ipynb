```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

```

```python

image = cv2.imread('sample.png')


image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)


plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)


blurred = cv2.GaussianBlur(gray, (5, 5), 0)


plt.imshow(blurred, cmap='gray')
plt.title("Grayscale with Gaussian Blur")
plt.axis("off")
plt.show()
```

```python

sobel_x = cv2.Sobel(blurred, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(blurred, cv2.CV_64F, 0, 1, ksize=3)

sobel_x_abs = cv2.convertScaleAbs(sobel_x)
sobel_y_abs = cv2.convertScaleAbs(sobel_y)
sobel_combined = cv2.addWeighted(sobel_x_abs, 0.5, sobel_y_abs, 0.5, 0)

plt.imshow(sobel_combined, cmap='gray')
plt.title("Sobel Edge Detection")
plt.axis("off")
plt.show()
```

```python
# Laplacian edge detection
laplacian = cv2.Laplacian(blurred, cv2.CV_64F)
laplacian_abs = cv2.convertScaleAbs(laplacian)

# Display Laplacian results
plt.imshow(laplacian_abs, cmap='gray')
plt.title("Laplacian Edge Detection")
plt.axis("off")
plt.show()
```

```python
# Apply Canny edge detection
canny_edges = cv2.Canny(blurred, 50, 150)

# Display Canny results
plt.imshow(canny_edges, cmap='gray')
plt.title("Canny Edge Detection")
plt.axis("off")
plt.show()
```

```python
# Plotting all results together
plt.figure(figsize=(12, 8))

# Original Image
plt.subplot(2, 3, 1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")

# Grayscale
plt.subplot(2, 3, 2)
plt.imshow(blurred, cmap='gray')
plt.title("Grayscale + Blur")
plt.axis("off")

# Sobel
plt.subplot(2, 3, 3)
plt.imshow(sobel_combined, cmap='gray')
plt.title("Sobel Edge Detection")
plt.axis("off")

# Laplacian
plt.subplot(2, 3, 4)
plt.imshow(laplacian_abs, cmap='gray')
plt.title("Laplacian Edge Detection")
plt.axis("off")

# Canny
plt.subplot(2, 3, 5)
plt.imshow(canny_edges, cmap='gray')
plt.title("Canny Edge Detection")
plt.axis("off")

plt.tight_layout()
plt.show()
```
